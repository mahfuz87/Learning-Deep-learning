```python
# import modules
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
np.random.seed(7)
```

```python
# load dataset
dataset = np.loadtxt("pima-indians-diabetes.csv", delimiter=",")
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
# create model
model = Sequential()
model.add(Dense(12, input_dim=8, kernel_initializer='uniform', activation='relu'))
model.add(Dense(8, kernel_initializer='uniform', activation='relu'))
model.add(Dense(1, kernel_initializer='uniform', activation='sigmoid'))

# this network has 8 inputs in the visible layer, 12 neuron in the first hidden layer, 8 neuron in the second hidden layer and 1 output in the output layer.

# we could initialize using gaussian distribution also. For gaussian distribution the keyword would be "normal" in stead of "uniform".
```

```python
# compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, Y, nb_epoch=450, batch_size=10, verbose=1)
```

```python
# evaluating the model
scores = model.evaluate(X,Y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
